```python
import numpy as np
```

```python
balls = 300
wickets = 10
state = np.array([balls,wickets])
```

```python
action = np.array([1,2,3,4,6])
```

```python
p_out_min = np.array([0.01,0.02,0.03,0.1,0.3])
```

```python
p_out_max = np.array([0.1,0.2,0.3,0.5,0.7])
```

```python
def agent_action(action_t,state,p_out_min,p_out_max):
        wickets = state[1]
        
        ind = np.where(action == action_t)
#         print(ind)
        x_out_max = p_out_max[ind[0]]
        x_out_min = p_out_min[ind[0]]
        
        p_out = x_out_max + (x_out_min - x_out_max) * ((wickets - 1)/9)
        
#         prob = np.random.choice([0,1],1,p=[p_out,1-p_out])
        
        prob = np.random.rand(1)

        score = 0
        
        if(prob <= p_out):
            state -= 1
            return np.array([score,state])
    
        else:
            state[0] -= 1
            
            y_run_max = 0.8
            y_run_min = 0.5
            
            p_run = y_run_min + (y_run_max - y_run_min) * ((wickets - 1)/9)
            
#             prob = np.random.choice([0,1],1,p=[p_run,1-p_run])
            
             prob = np.random.rand(1)
    
            if prob <= p_run:
                score = action_t
            else:
                score = 0
            return np.array([score,state])
            
    
```

```python
act = 1

global  first,last
first = 0
last = 0
count = 0


while state[1] != 0:

    count += 1
    ball = agent_action(act,state,p_out_min,p_out_max)
    print(ball)
    x = ball[1]
    
    if x[1] == 10:
        first += 1
    
    if x[1] == 1:
        last += 1
        
    state = x
    print(state)
        
print(first,last)

print(first/count , last/count)
        



```
