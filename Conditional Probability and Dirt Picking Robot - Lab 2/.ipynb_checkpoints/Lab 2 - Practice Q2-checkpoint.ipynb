```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
xsize = 7
ysize = 7

agent_x = np.random.randint(0,xsize,1)
agent_y = np.random.randint(0,ysize,1)

def createRoom(x,y):
    global room
    room = np.zeros((x,y),dtype='int')
    return room
```

```python
def createDirt(room):
    x = np.random.randint(0,xsize,1)
    y = np.random.randint(0,ysize,1)
    print( "Dirt position : {},{}".format(x[0],y[0]))
    room[x,y] += 1
    return room
    
```

```python
def agent_action():
    return np.random.randint(0,5,1)
```

```python
def reward(action,dirtRoom,hit):
     
    expense = 0
    if(hit == 1):
        expense = -10
    
    if(action == 4 and dirtRoom[agent_x,agent_y]==0):
        expense = -1
    
   
    reward = dirtRoom[agent_x,agent_y]
    reward += expense
    
    return reward
```

```python
room = createRoom(xsize,ysize)
static_room = np.copy(room)
perf = 0
for r in range(100):
    print("Agent position: {},{}".format(agent_x,agent_y))
    static_room[agent_x,agent_y]=1
    print(static_room)
    static_room[agent_x,agent_y]=0

    dirtarray = createDirt(room)
    print(dirtarray)
    
    action = agent_action()
    hit = 0
    
    if(action == 0):
        print('Agent Action: Up')
        move = agent_x - 1
        if(move > xsize-1 or move < 0 ):
            hit = 1
        else:
            agent_x -= 1
            hit = 0
    
    elif (action == 1):
        print('Agent Action: Down')
        move = agent_x + 1
        if(move > xsize-1 or move < 0 ):
            hit = 1
        else:
            hit = 0
            agent_x += 1
        
    elif (action == 2):
        print('Agent Action: Left')
        move = agent_y - 1
        if(move > ysize-1 or move < 0 ):
            hit = 1
        else:
            hit = 0
            agent_y -= 1
     
    elif(action == 3):
        print('Agent Action: Right')
        move = agent_y + 1
        if(move > ysize-1 or move < 0 ):
            hit = 1
        else:
            hit = 0
            agent_y += 1
        
    elif(action == 4):
        print('Agent Action: Pick Up Dirt')
        hit = 0
        
    re = reward(action,dirtarray,hit)
    perf += re
    print("Reward: {}".format(re))
    
    print('\n')
    
    
```
