```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
plt.style.use('seaborn')
```

```python
class gradientDescent:
    
    def __init__(self,a,b,c,alpha):
        
        self.a = a
        self.b = b
        self.c = c
        
        self.weight = np.random.uniform(-1000,1000,1)
        
        self.ws = np.linspace(-1000,1000,1000)
        
        self.lw = lambda x : (a/2)*np.power(x,2) + b*x + c
        self.diff = lambda x : a * x + b
        self.lr = alpha
        
    def runDescent(self):
        
        
        w_star = -self.b/self.a
        
        for t in range(1000):
            
            self.weight -= self.lr * self.diff(self.weight)
            
            plt.plot(self.ws,self.lw(self.ws))
#             plt.plot(self.weight - w_star,self.lw(self.weight - w_star),'ob')
            
           
            plt.plot(self.weight,self.lw(self.weight),'ob')
            
            
            
            
     
    def plotline(self):
        
        plt.plot(self.ws,self.lw(self.ws),'r')
        w_star = -self.b/self.a
        plt.plot(w_star,30,'og')
        plt.plot(-self.b/self.a,self.lw(w_star),'ob')
            
    
    
    
    
    
```

```python
algo = gradientDescent(1,0,0,1.5)
```

```python
algo.plotline()
```

```python
algo.runDescent()
```
