```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
class gradientDescent2D:
    
    def __init__(self,lr,a,b):
        
        self.weights = 2 * np.random.rand(2) - 1
        self.lr = lr
        self.a = a
        self.b = b
        self.lossf = lambda x,y : (a)*np.power(x,2) + (b)*np.power(y,2)
        self.tol = 0.00001       
        
    def p_diff(self,x,y):
        
        loss_w1 = 2*self.a*x
        loss_w2 = 2*self.b*y
        
        return [loss_w1,loss_w2]
        
        
        
        
    def runDescent(self):
        
        diff = self.p_diff(self.weights[0],self.weights[1])
        
        stepsize_w1 = self.lr * diff[0]
        stepsize_w2 = self.lr * diff[1]
        
        while abs(stepsize_w1) > self.tol or abs(stepsize_w2) > self.tol:
        
        
            if(abs(stepsize_w1) > self.tol):                
                self.weights[0] -= stepsize_w1
                
            if(abs(stepsize_w2) > self.tol):
                self.weights[1] -= stepsize_w2
            
            diff = self.p_diff(self.weights[0],self.weights[1])
        
            stepsize_w1 = self.lr * diff[0]
            stepsize_w2 = self.lr * diff[1]
            print(self.lossf(self.weights[0],self.weights[1]))
            
            plt.plot(self.weights[0],self.weights[1],'ob')
            

        
    
    
        self.createPlot()
    
    def createPlot(self):
        
        
        xlist = np.linspace(-5, 5, 10)
        ylist = np.linspace(-5, 5, 10)
        X, Y = np.meshgrid(xlist, ylist)
        Z = self.a * np.power(X,2) + self.b * np.power(Y,2)
        
        
        
        plt.quiver(X,Y,-self.p_diff(X,Y)[0],-self.p_diff(X,Y)[1],self.lossf(X,Y))
        
        
        plt.contour(X,Y,Z)
    
    
    
    
    
    
    
```

```python
opt = gradientDescent2D(0.0001,0.5,0.5)
```
